```python
import pandas as pd
from prophet import Prophet
from sklearn.preprocessing import LabelEncoder


# Carregamento dos dados
df = pd.read_csv('database/combined_sales.csv')

# Converter a coluna de data para o tipo correto
df['date'] = pd.to_datetime(df['date'])

# Codificar os nomes das categorias para números
le = LabelEncoder()
df['item_category_id'] = le.fit_transform(df['item_category_name'])

# Criar um dataframe para cada categoria de item
df_grouped = df.groupby('item_category_id')

# Treinar um modelo para cada categoria
models = {}
for category, group in df_grouped:
    group = group.rename(columns={'date': 'ds', 'item_cnt_day': 'y'})
    group = group.groupby(pd.Grouper(key='ds', freq='M')).sum().reset_index()
    
    if group['y'].notna().sum() < 2:
        print(f'Not enough data for category {category}, skipping...')
        continue
    
    m = Prophet(yearly_seasonality=True)
    m.fit(group)
    models[category] = m


# Função para prever vendas
def predict_sales(category_name, month):
    category_id = le.transform([category_name])[0]
    future = models[category_id].make_future_dataframe(periods=month)
    forecast = models[category_id].predict(future)
    return forecast
```

```python
print(predict_sales('Кино - Blu-Ray', 1))
```
